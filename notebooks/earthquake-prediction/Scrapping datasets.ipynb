```
import pandas as pd
```

```
def format_path(_from, _to):
    return 'https://earthquake.usgs.gov/fdsnws/event/1/query.csv?starttime=' + _from + '%2000:00:00&endtime=' + _to + '%2000:00:00&orderby=time-asc'

def format_year(year):
    return str(year)

def format_month(month):
    return '0' + str(month) if month < 10 else str(month)

def format_day(day):
    return '0' + str(day) if day < 10 else str(day)

def get_earthquake_dataset(year):

    _month_from = 1
    _month_to = 13
    _month_diff = 1
    _day_from = 1
    _day_to = 22
    _day_diff = 10

    # 2020-01-01 to 2020-01-11
    # 2020-01-11 to 2020-01-21
    # 2020-01-21 to 2020-02-01
    # When it is december (12)
    # 2020-12-21 to 2021-01-01

    for idx in range(_month_from, _month_to, _month_diff):
        for idy in range(_day_from, _day_to, _day_diff):
            month = format_month(idx)
            day = format_day(idy)

            if idx == 1 and idy == 1:
                # For baseline, we make range for 10 days
                next_day = format_day(idy + _day_diff)
                _from = f'{format_year(year)}-{month}-{day}'
                _to = f'{format_year(year)}-{month}-{next_day}'
                # Read dataset and save to df
                df = pd.read_csv(format_path(_from, _to))
                print(f'Append data {_from} > {_to} ===== {len(df)}')

            elif idx != 12:
                if idy != 21:
                    next_day = format_day(idy + _day_diff)
                    _from = f'{format_year(year)}-{month}-{day}'
                    _to = f'{format_year(year)}-{month}-{next_day}'
                    # Read dataset and save to df
                    temp = pd.read_csv(format_path(_from, _to))
                    df = df.append(temp)
                    print(f'Append data {_from} > {_to} ===== {len(df)}')
                else:
                    next_day = format_day(1)
                    next_month = format_month(idx + _month_diff)
                    _from = f'{format_year(year)}-{month}-{day}'
                    _to = f'{format_year(year)}-{next_month}-{next_day}'
                    temp = pd.read_csv(format_path(_from, _to))
                    df = df.append(temp)
                    print(f'Append data {_from} > {_to} ===== {len(df)}')
            else:
                if idy != 21:
                    next_day = format_day(idy + _day_diff)
                    _from = f'{format_year(year)}-{month}-{day}'
                    _to = f'{format_year(year)}-{month}-{next_day}'
                    # Read dataset and save to df
                    temp = pd.read_csv(format_path(_from, _to))
                    df = df.append(temp)
                    print(f'Append data {_from} > {_to} ===== {len(df)}')
                else:
                    next_day = format_day(1)
                    next_month = format_month(1)
                    next_year = format_year(year + 1)
                    _from = f'{format_year(year)}-{month}-{day}'
                    _to = f'{format_year(next_year)}-{next_month}-{next_day}'
                    temp = pd.read_csv(format_path(_from, _to))
                    df = df.append(temp)
                    print(f'Append data {_from} > {_to} ===== {len(df)}')

    print(f'Lenght of data {year}: {len(df)} =====')
    filename = f'{format_year(year)}.csv'
    df.sort_values('time', ascending=True).reset_index()
    df.to_csv(filename)
    #return df
```

```
# 1900-1984
# 1985-1999
# 2000-2001
# 2002-2019
for y in range(1900, 2020):
    get_earthquake_dataset(y)
```
